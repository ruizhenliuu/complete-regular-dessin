# Isogeny computations

```python
%display latex
x,y = var('x y')
alpha=sqrt(-2)
```

```python
xp(x,y)= alpha^(-2)*(x+4+ 2/x)
yp(x,y)=alpha^(-3)*y*(1- 2/(x^2))
lmd(x,y) = simplify((yp-y)/(xp-x))
nu(x,y) = simplify((y*xp-yp*x)/(xp-x))
```

```python
xr(x,y)=simplify(lmd^2-4-x-xp)
yr(x,y) = simplify(-lmd*xr - nu)
xr,yr
```
